# Filtrado
En base al anlisis efectuado hasta aqu, se propone realizar los siguientes cambios: Por un lado, eliminar las columnas 'oldbalanceOrg', 'newbalanceOrig', 'oldbalanceDest', 'newbalanceDest', 'isFlaggedFraud' debido a la escasa relevancia de estos datos. Y por otro lado, utilizar los datos cuyo 'type' sea 'TRANSFER' o 'CASH_OUT', ya que por su naturaleza son las nicas transacciones donde se ha detectado fraude.


```python
df_filtrado = df[['step', 'type', 'amount', 'nameOrig',
                  'nameDest', 'isFraud']]

df_filtrado = df_filtrado[(df_filtrado['type']=='TRANSFER') | (df_filtrado['type']=='CASH_OUT')]

df_filtrado.head(10)
```
